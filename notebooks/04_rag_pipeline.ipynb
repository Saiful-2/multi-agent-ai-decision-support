```python
# ============================================================
# CELL 06
# LOAD VECTOR DATABASE
# ============================================================

index = faiss.read_index(

    "faiss_index.bin"

)

texts = joblib.load(

    "texts.pkl"

)

print(

    "Vector Database Loaded"

)

print(

    "Documents:",

    len(texts)

)
```

```python
# ============================================================
# CELL 07
# LOAD EMBEDDING MODEL
# ============================================================

MODEL_NAME = (

    "all-MiniLM-L6-v2"

)

embedding_model = (

    SentenceTransformer(

        MODEL_NAME

    )

)

print(

    "Embedding model loaded."

)
```

```python
# ============================================================
# CELL 08
# GEMINI API CONFIGURATION
# ============================================================

GOOGLE_API_KEY = "AIzaSyBI_IaNPa2-HlVqp1xzoM2grtIfQZAFIf4"

genai.configure(
    api_key=GOOGLE_API_KEY
)

llm = genai.GenerativeModel(
    "gemini-2.5-flash"
)

print(
    "Gemini connected."
)
```

```python
# ============================================================
# CELL 09
# QUESTION INPUT
# ============================================================

query = input(

    "Enter Question: "

)
```

```python
# ============================================================
# CELL 10
# QUERY EMBEDDING
# ============================================================

query_vector = (

    embedding_model.encode(

        [query],

        convert_to_numpy=True

    )

)
```

```python
# ============================================================
# CELL 11
# VECTOR SEARCH
# ============================================================

TOP_K = 5

distances, indices = (

    index.search(

        query_vector,

        TOP_K

    )

)
```

```python
# ============================================================
# CELL 12
# RETRIEVE CONTEXT
# ============================================================

retrieved_context = []

for idx in indices[0]:

    retrieved_context.append(

        texts[idx]

    )

context = "\n\n".join(

    retrieved_context

)

print(

    context[:1500]

)
```

```python
# ============================================================
# CELL 13
# CREATE PROMPT
# ============================================================

prompt = f"""

You are an intelligent
decision support AI.

Answer ONLY using
provided context.

If answer unavailable,
say:

"I do not have enough information."

CONTEXT:

{context}

QUESTION:

{query}

ANSWER:

"""
```

```python
# ============================================================
# CELL 14
# GENERATE RESPONSE
# ============================================================

try:
    response = llm.generate_content(
        prompt
    )

    answer = response.text

    print(
        answer
    )
except Exception as e:
    print(f"An error occurred during content generation: {e}")
    print("Listing available models to diagnose the issue...")
    for m in genai.list_models():
        if "generateContent" in m.supported_generation_methods:
            print(f"  - {m.name}")
```

```python
# ============================================================
# CELL 15
# FULL PIPELINE FUNCTION
# ============================================================

def rag_pipeline(

    question,

    top_k=5

):

    vector = (

        embedding_model.encode(

            [question],

            convert_to_numpy=True

        )

    )

    D,I = index.search(

        vector,

        top_k

    )

    context = "\n\n".join(

        [

            texts[idx]

            for idx

            in I[0]

        ]

    )

    prompt = f"""

    Context:

    {context}

    Question:

    {question}

    Answer:

    """

    response = (

        llm.generate_content(

            prompt

        )

    )

    return response.text
```
