```python
# ============================================================
# CELL 03
# IMPORT LIBRARIES
# ============================================================

import random
import numpy as np
import pickle

from google.colab import files

from langchain_community.document_loaders import (
    PyPDFLoader
)

from langchain_text_splitters import (
    RecursiveCharacterTextSplitter
)

print("Libraries loaded successfully.")
```

```python
# ============================================================
# CELL 04
# RANDOM SEED
# ============================================================

SEED = 42

random.seed(SEED)

np.random.seed(SEED)

print(
    "Random seed fixed:",
    SEED
)
```

```python
# ============================================================
# CELL 05
# UPLOAD PDF FILES
# ============================================================

uploaded = files.upload()
```

```python
# ============================================================
# CELL 06
# CHECK FILES
# ============================================================

uploaded_files = list(
    uploaded.keys()
)

print(
    "Uploaded files:"
)

for file in uploaded_files:

    print(file)
```

```python
# ============================================================
# CELL 07
# LOAD PDF DOCUMENTS
# ============================================================

documents = []

for file in uploaded_files:

    loader = PyPDFLoader(file)

    pages = loader.load()

    documents.extend(pages)

print(
    "Total Pages:",
    len(documents)
)
```

```python
# ============================================================
# CELL 08
# PREVIEW DOCUMENT
# ============================================================

documents[0].page_content[:1000]
```

```python
# ============================================================
# CELL 09
# CREATE CHUNKER
# ============================================================

CHUNK_SIZE = 1000

CHUNK_OVERLAP = 200

splitter = RecursiveCharacterTextSplitter(

    chunk_size=CHUNK_SIZE,

    chunk_overlap=CHUNK_OVERLAP

)

print(

    "Chunk Size:",

    CHUNK_SIZE

)

print(

    "Chunk Overlap:",

    CHUNK_OVERLAP

)
```

```python
# ============================================================
# CELL 10
# CREATE CHUNKS
# ============================================================

chunks = splitter.split_documents(
    documents
)

print(

    "Total Chunks:",

    len(chunks)

)
```

```python
# ============================================================
# CELL 11
# PREVIEW CHUNK
# ============================================================

chunks[0].page_content
```

```python
# ============================================================
# CELL 12
# CHUNK LENGTH STATISTICS
# ============================================================

chunk_lengths = [

    len(
        chunk.page_content
    )

    for chunk in chunks

]

print(

    "Minimum:",

    min(chunk_lengths)

)

print(

    "Maximum:",

    max(chunk_lengths)

)

print(

    "Average:",

    round(

        np.mean(
            chunk_lengths
        ),

        2

    )

)
```

```python
# ============================================================
# CELL 13
# SAVE CHUNKS
# ============================================================

with open(

    "chunks.pkl",

    "wb"

) as file:

    pickle.dump(

        chunks,

        file

    )

print(

    "Chunks saved."

)
```

```python
# ============================================================
# CELL 14
# DOWNLOAD CHUNKS
# ============================================================

files.download(
    "chunks.pkl"
)
```
