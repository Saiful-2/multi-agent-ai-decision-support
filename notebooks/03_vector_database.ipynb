```python
# ============================================================
# CELL 03
# IMPORT LIBRARIES
# ============================================================

import random

import numpy as np

import joblib

import faiss

from google.colab import files

print(

    "Libraries loaded successfully."

)
```

```python
# ============================================================
# CELL 04
# RANDOM SEED
# ============================================================

SEED = 42

random.seed(SEED)

np.random.seed(SEED)

print(

    "Random seed fixed:",

    SEED

)
```

```python
# ============================================================
# CELL 05
# UPLOAD FILES
# ============================================================

uploaded = files.upload()
```

```python
# ============================================================
# CELL 06
# LOAD FILES
# ============================================================

embeddings = joblib.load(

    "embeddings.pkl"

)

texts = joblib.load(

    "texts.pkl"

)

print(

    "Embeddings Shape:",

    embeddings.shape

)

print(

    "Texts Loaded:",

    len(texts)

)
```

```python
# ============================================================
# CELL 07
# EMBEDDING DIMENSION
# ============================================================

dimension = embeddings.shape[1]

print(

    "Vector Dimension:",

    dimension

)
```

```python
# ============================================================
# CELL 08
# CREATE FAISS INDEX
# ============================================================

index = faiss.IndexFlatL2(

    dimension

)

print(

    "FAISS Index Created"

)
```

```python
# ============================================================
# CELL 09
# ADD EMBEDDINGS
# ============================================================

index.add(

    embeddings

)

print(

    "Vectors Added:",

    index.ntotal

)
```

```python
# ============================================================
# CELL 10
# VECTOR DATABASE INFORMATION
# ============================================================

print(

    "Database Size:",

    index.ntotal

)

print(

    "Dimension:",

    dimension

)
```

```python
# ============================================================
# CELL 11
# SAMPLE QUERY
# ============================================================

query = """

What is machine learning?

"""

print(

    query

)
```

```python
# ============================================================
# CELL 12
# LOAD EMBEDDING MODEL
# ============================================================

from sentence_transformers import (

    SentenceTransformer

)

MODEL_NAME = (

    "all-MiniLM-L6-v2"

)

model = SentenceTransformer(

    MODEL_NAME

)

print(

    "Model Loaded"

)
```

```python
# ============================================================
# CELL 13
# QUERY EMBEDDING
# ============================================================

query_embedding = model.encode(

    [query],

    convert_to_numpy=True

)
```

```python
# ============================================================
# CELL 14
# VECTOR SEARCH
# ============================================================

TOP_K = 5

distances, indices = index.search(

    query_embedding,

    TOP_K

)
```

```python
# ============================================================
# CELL 15
# SEARCH RESULT INDICES
# ============================================================

print(

    "Indices"

)

print(

    indices

)

print(

    "Distances"

)

print(

    distances

)
```

```python
# ============================================================
# CELL 16
# RETRIEVED DOCUMENTS
# ============================================================

for i, idx in enumerate(

    indices[0]

):

    print(

        "="*60

    )

    print(

        "Rank:",

        i+1

    )

    print(

        texts[idx][:500]

    )
```

```python
# ============================================================
# CELL 17
# SAVE VECTOR DATABASE
# ============================================================

faiss.write_index(

    index,

    "faiss_index.bin"

)

print(

    "FAISS saved."

)
```

```python
# ============================================================
# CELL 18
# DOWNLOAD VECTOR DATABASE
# ============================================================

files.download(

    "faiss_index.bin"

)
```

```python
# ============================================================
# CELL 19
# DATABASE SUMMARY
# ============================================================

summary = {

    "Vectors":index.ntotal,

    "Dimension":dimension,

    "Top_K":TOP_K

}

summary
```
